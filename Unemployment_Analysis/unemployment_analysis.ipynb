```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dp=pd.read_csv("Unemployment.csv")
dp.head()
```

```python
dp.info()
```

```python
dp.columns = dp.columns.str.strip()
```

```python
dp["Date"] = pd.to_datetime(dp["Date"])
dp = dp.sort_values("Date")
```

```python
plt.figure()
```

```python
sns.lineplot(data=dp,x = "Date" , y = "Estimated Unemployment Rate (%)")
plt.title("Unemployment Rate Over time in India")
plt.xticks(rotation=45)
plt.show()
```

```python
region_avg = dp.groupby("Region")["Estimated Unemployment Rate (%)"].mean().sort_values()
plt.figure()
region_avg.plot(kind="barh")
plt.title("Average Unemployment Rate by Region in India")
plt.show()
```
